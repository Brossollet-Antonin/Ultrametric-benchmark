```python
os.chdir(cwd)

%matplotlib inline
import sequence_generator_temporal as sequence_generator
import numpy as np
from matplotlib import pyplot as plt

import control
```

```python
tree_depth = 7
branching = 2
n_classes = branching**tree_depth
```

```python
temp_set = [1/1.2, 1/1.4, 1/1.8, 1/2, 1/2.2, 1.5, 2, 1/3]
cors = [[] for i in range(len(temp_set))]
sequences = [[] for i in range(len(temp_set))]

seqgen = sequence_generator.TempCorr_SequenceGenerator()

for i,T in enumerate(temp_set):
    sequences[i] = seqgen.generate_labels(
        sequence_first = 0,
        sequence_length = 10000,
        energy_step = 1,
        T = T,
        tree_depth = tree_depth,
        tree_branching = branching,
        verbose = False
    )[0]
    cors[i] = sequence_generator.sequence_autocor(sequences[i], n_classes, nlags=500)
    plt.plot(cors[i], label=T)
    plt.xscale("log")
    plt.yscale("log")
    
plt.legend()

plt.figure()
plt.plot(sequences[0])
plt.figure()
plt.plot(sequences[5])    
plt.figure()
plt.plot(sequences[2])
```

```python
average = 9
T=1/4

seq = seqgen.generate_labels(
        sequence_first = 0,
        sequence_length = 10000,
        energy_step = 1,
        T = T,
        tree_depth = tree_depth,
        tree_branching = branching,
        verbose = False
    )[0]

cor_avg =  np.array(sequence_generator.sequence_autocor(seq, n_classes))

for i in range(average):
    sequence_ = np.array(
        seqgen.generate_labels(
            sequence_first = 0,
            sequence_length = 10000,
            energy_step = 1,
            T = T,
            tree_depth = tree_depth,
            tree_branching = branching,
            verbose = False
        )[0]
    )
    cor_avg = cor_avg + sequence_generator.sequence_autocor(sequence_, n_classes)
    
cor_avg = cor_avg/10

plt.figure()

plt.plot(cor_avg)
```

```python
average = 9
T=1/4
block_size = 500

cor_avg_block =  np.array(
    sequence_generator.sequence_autocor(
        control.shuffle_block(
            seqgen.generate_labels(
                sequence_first = 0,
                sequence_length = 10000,
                energy_step = 1,
                T = T,
                tree_depth = tree_depth,
                tree_branching = branching,
                verbose = False
            )[0], block_size
        ), n_classes
    )
)

for i in range(average):
    sequence_ = np.array(
        sequence_generator.sequence_autocor(
            control.shuffle_block(
                seqgen.generate_labels(
                    sequence_first = 0,
                    sequence_length = 10000,
                    energy_step = 1,
                    T = T,
                    tree_depth = tree_depth,
                    tree_branching = branching,
                    verbose = False
                )[0], block_size
            ), n_classes
        )
    )
    cor_avg_block = cor_avg_block + sequence_generator.sequence_autocor(sequence_, n_classes)
    
cor_avg_block = cor_avg_block/10

plt.figure()
plt.plot(cor_avg_block)
```
